## Part 1: Data Generation

We will generate a synthetic dataset simulating social media posts with a rich schema.

**Schema**:
- `post_id` (String): Unique identifier.
- `user_id` (Integer): User identifier.
- `timestamp` (DateTime): Time of post.
- `content` (String): Text content.
- `likes` (Integer): Number of likes.
- `views` (Integer): Number of views.
- `category` (String): Post category.
- `tags` (List[String]): Hashtags.
- `location` (String): User location.
- `device` (String): Device used (Mobile, Web, etc.).
- `latency` (Float): Network latency.
- `error_rate` (Float): Error rate during upload.

```python
def generate_data(num_records=1_000_000, output_path="social_media_data.parquet"):
    fake = Faker()
    
    print(f"Generating {num_records} records...")
    
    # Generate data using numpy for speed where possible
    data = {
        "post_id": [fake.uuid4() for _ in range(num_records)],
        "user_id": np.random.randint(1, 100_000, num_records),
        "timestamp": pd.date_range(start="2023-01-01", periods=num_records, freq="s").to_numpy().astype("datetime64[us]"),
        "likes": np.random.randint(0, 10_000, num_records),
        "views": np.random.randint(0, 1_000_000, num_records),
        "category": np.random.choice(["Tech", "Health", "Travel", "Food", "Fashion", "Politics", "Sports"], num_records),
        "tags": [np.random.choice(["#viral", "#new", "#trending", "#hot", "#update"], size=np.random.randint(1, 4)).tolist() for _ in range(num_records)],
        "location": np.random.choice(["USA", "UK", "DE", "PL", "FR", "JP", "BR"], num_records),
        "device": np.random.choice(["Mobile", "Desktop", "Tablet"], num_records),
        "latency": np.random.uniform(10.0, 500.0, num_records),
        "error_rate": np.random.beta(1, 10, num_records),
        "content": [fake.sentence() for _ in range(min(num_records, 1000))] * (num_records // 1000 + 1)
    }
    
    # Trim to exact size
    data["content"] = data["content"][:num_records]
    
    df = pd.DataFrame(data)
    
    print("Writing to Parquet...")
    df.to_parquet(output_path, engine="pyarrow")
    print(f"Data saved to {output_path}")

# Generate 5 million records
# generate_data(num_records=5_000_000)
```

```python
TRIES = 100
from functools import partial
%load_ext memory_profiler

def profile_time_and_memory(func, tries=100, *args, **kwargs):
    mem_usage = memory_usage((func, args, kwargs), max_usage=True, retval=True)
    time = timeit.timeit(partial(func, *args, **kwargs), number=tries) / tries
    print(f"Time: {time} seconds")
    print(f"Peak Memory: {mem_usage[0]} MiB")
    return ""
```

```python
print("query C - Join:")

def f(df):
    # 1. Filter the source dataframes (WHERE clause)
    # df1: views > 500000
    df1_filtered = df[df['views'] > 500000]
    # df2: likes > 2
    df2_filtered = df[df['likes'] > 2][:200_000]

    # 2. Perform the self-join (FROM ... JOIN ... ON)
    # We join on 'likes'. This keeps rows where likes match.
    merged = pd.merge(df1_filtered, df2_filtered, on='likes', suffixes=('', '_df2'))

    # 3. Apply the aggregation filter (HAVING)
    # Condition: df1.likes + df2.likes > 1000
    # Since we joined on 'likes', df1.likes and df2.likes are the same value.
    merged = merged[merged['likes'] + merged['likes'] > 1000]

    # 4. Group by and Select (GROUP BY ... SELECT)
    # We get distinct pairs of likes. Since they are identical, we just need unique 'likes'.
    result = merged[['likes']].drop_duplicates()

    # 5. Format output to match SQL "SELECT df1.likes, df2.likes"
    result['df1_likes'] = result['likes']
    result['df2_likes'] = result['likes']
    final_output = result[['df1_likes', 'df2_likes']]

    print(final_output)

profile_time_and_memory(partial(f, df_pd), 5)
```

# Polars

```python
print("query A - Mean Latency:")
profile_time_and_memory(df_pl["latency"].mean)
```

```python
print("query B - Mean Latency for each user:")
profile_time_and_memory(lambda: df_pl.group_by("user_id").agg(pl.col("latency").mean()))
```

```python
print("query C - Join:")

profile_time_and_memory(lambda: (
    df_pl.lazy()
    # 1. Filter df1 (WHERE df1.views > 500000)
    .filter(pl.col("views") > 500_000)
    .join(
        # 2. Filter df2 (WHERE df2.likes > 2) and Join
        df_pl.lazy().limit(200_000).filter(pl.col("likes") > 2),
        on="likes",
        how="inner"
    )
    # 3. Apply Aggregation Filter (HAVING df1.likes + df2.likes > 1000)
    # Since we joined on 'likes', the value is identical for both
    .filter((pl.col("likes") * 2) > 1000)
    # 4. Select and Group By (DISTINCT)
    .select(
        pl.col("likes").alias("df1_likes"),
        pl.col("likes").alias("df2_likes")
    )
    .unique()
    .collect()
))
```

```python
# Query C: Join
print("query C - Join:")
STMT = """
    SELECT df1.likes, df2.likes
    FROM 'social_media_data.parquet' AS df1
       JOIN 'social_media_data.parquet' AS df2 ON (df1.likes = df2.likes)
    WHERE df1.views > 500000 AND df2.likes > 2
    GROUP BY df1.likes, df2.likes
    HAVING df1.likes + df2.likes > 1000
"""
profile_time_and_memory(partial(duckdb.sql, STMT))
```

```python
print("query A - Mean Latency:")
profile_time_and_memory(partial(df_spark.selectExpr, "mean(latency)"))
```

```python
print("query B - Mean Latency for each user:")
profile_time_and_memory(partial(df_spark.groupBy("user_id").avg, "latency")
```

```python
print("query C - Join:")
from pyspark.sql import functions as F

def fun(df):
    df1 = df.alias("df1")
    df2 = df.alias("df2")
    return (df1
        .filter(F.col("df1.views") > 500000)
        .join(
            df2.filter(F.col("df2.likes") > 2),
            F.col("df1.likes") == F.col("df2.likes"),
            "inner"
        )
        .groupBy(F.col("df1.likes"), F.col("df2.likes"))
        .agg(
            F.first("df1.likes").alias("likes_1"),
            F.first("df2.likes").alias("likes_2")
        )
        .filter(F.col("likes_1") + F.col("likes_2") > 1000)
        .select(
            F.col("likes_1").alias("likes"),
            F.col("likes_2").alias("likes")
        )
    )

profile_time_and_memory(partial(fun, df_spark))
```
